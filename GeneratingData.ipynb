```python
R = np.random.randn(1000, 3)  # Example usage of numpy
# This code generates a 1000x3 array of random numbers from a standard normal distribution using numpy.
```

```python
R.mean(axis=0)  # Example usage of numpy
```

```python
R.mean(axis=1)  # Example usage of numpy
# This code calculates the mean of each row in the array R, which contains random numbers from a standard normal distribution.
```

```python
R.mean(axis = 1).shape  # Example usage of numpy
# This code returns the shape of the array obtained by calculating the mean of each row in the array R, which contains random numbers from a standard normal distribution.
# The shape will be (1000,) since it calculates the mean across the columns for each row.
```

```python
np.cov(R)  # Example usage of numpy
# This code calculates the covariance matrix of the array R, which contains random numbers from a standard normal distribution.
# The covariance matrix provides insights into the relationships between the different columns of the array.
```

```python
np.cov(R).shape  # Example usage of numpy
# This code returns the shape of the covariance matrix calculated from the array R, which contains random numbers from a standard normal distribution.
```

```python
np.cov(R.T)  # Example usage of numpy
# This code calculates the covariance matrix of the transposed array R, which contains random numbers from a standard normal distribution.
```

```python
np.cov(R, rowvar=False)  # Example usage of numpy
# This code calculates the covariance matrix of the array R, treating each column as a variable and each row as an observation.
# The `rowvar=False` argument indicates that the rows represent observations rather than variables.
# This is useful when the data is structured such that each column represents a different variable, and each row represents a different observation.
# The covariance matrix provides insights into the relationships between the different variables in the dataset.
```
