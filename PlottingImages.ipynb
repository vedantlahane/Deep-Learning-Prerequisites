```python
gray = arr.mean(axis = 2)  # Convert to grayscale by averaging the RGB channels
```

```python
gray.shape  # This will give you the dimensions of the grayscale image (height, width)
```

```python
plt.imshow(gray)  # Display the grayscale image using matplotlib
```

```python
plt.imshow(gray, cmap='gray')  # Display the grayscale image using matplotlib
plt.axis('off')  # Hide the axes
```
