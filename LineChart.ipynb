```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1000)  # Generate 100 points between 0 and 10
```

```python
y = np.sin(x) + 0.2 *  x  # Generate y values using a sine function with some noise
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('Line Chart Example')
```

```python
plt.plot(x, y, label='Sine Wave with Noise')  # Plot the line chart
plt.xlabel('X-axis')  # Label for the x-axis
plt.ylabel('Y-axis')  # Label for the y-axis
plt.title('Line Chart Example');  # Title of the chart
# ; is used to suppress the output in Jupyter notebooks
```
