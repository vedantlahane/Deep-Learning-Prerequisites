```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(100 ,2)  # Generate 100 random points in 2D
plt.scatter(X[:, 0], X[:, 1]) # Scatter plot of the points  
```

```python
X = np.random.randn(200 ,2)
X[:50] += 3 # Shift the first 50 points to create two clusters
Y = np.zeros(200) # Create labels for the points
Y[:50] = 1 # Assign label 1 to the first 50 points
plt.scatter(X[:, 0], X[:, 1], c=Y);
```
