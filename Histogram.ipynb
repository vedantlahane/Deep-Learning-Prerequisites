```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(10000)  # Generate 10,000 random points from a normal distribution
```

```python
plt.hist(X);
```

```python
plt.hist(X,bins=50); # Specify the number of bins for the histogram , bins are the intervals into which the data is divided
```

```python
X = np.random.random(10000)
plt.hist(X, bins=50, density=True);  # Density=True normalizes the histogram , that means the area under the histogram will sum to 1
```
